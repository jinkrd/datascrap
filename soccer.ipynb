```python
import pandas as pd
data_frame=pd.read_csv('WorldCups.csv')
data_frame1=pd.DataFrame(data_frame,columns=['Year','Winner','GoalsScored'])
data_frame1
```

```python
import seaborn as sns
sns.set()

graph = sns.scatterplot(x='Year',y='GoalsScored',data=data_frame1)
graph
```

```python
from bokeh.plotting import figure,show
from bokeh.models import HoverTool

TOOLTIPS = HoverTool(tooltips=[
    ("index","$index"),
    ("(Year,GoalsScored)","(@Year,@GoalsScored)"),
    ("Winner","@Winner")]
)
p=figure(title="soccer",x_axis_label='Year',y_axis_label='GoalsScored',plot_width=200,plot_height=500,tools=[TOOLTIPS])
p.circle('Year','GoalsScored',size=10,source=data_frame1)
show(p)
```
